```python
# import libraries and modules

import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
sn.set()
```

```python
# point to the data source
data_source = "../../../1.04. Real-life example.csv"

# read the data into
data = pd.read_csv(data_source)

# Examiine the first few column of the data

data.head(10)
```

### Preprocessing

```python
# perform descriptive statistics

data.describe(include = "all")
```

```python
# For this regression, the MOdel column will be drop

data.drop("Model", axis = 1, inplace = True)

# checking the data description one more time 
data.describe(include = "all")
```

### dealing with missing value

```python
data.isnull().sum()
```

```python
# from the above, Price and EngineV has a lot of missing values

# to remove rows where there are missing value

data.dropna(axis = 0, inplace = True)

# checking again for missing values

data.isnull().sum()
```

```python
# check the descriptive stat without the missing value

data.describe(include = "all")
```

## Exploring the Probability disribution function

```python
sn.distplot(data["Price"])
plt.title("Price Distribution With Outlier");
```

```python
# the price column has a lot of outlier which can affect the result of the regression 
# to remove the outlier
# find the 0.99  for the price column
Q = data["Price"].quantile(0.99)

# retain only data where qualtile is below the 0.99 quantile
data_1 = data[data["Price"]<Q]

data_1.describe(include = "all")
```

```python
# let's now check the ditribution of the price column one more time

sn.distplot(data_1["Price"])
plt.title("Price Distribution afte outlier removed");
```

```python
# let's check for outlier in Mileage column

sn.distplot(data_1["Mileage"])
plt.title("Mileage Distribution with Outlier");
```
